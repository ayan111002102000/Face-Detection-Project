```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
```

```python
# load the pre-trained model
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# load the photograph
sample_image = cv2.imread('image1.jpg')
```

```python
plt.imshow(plt.imread('image1.jpg'))
```

```python
# perform face detection
bboxes = face_detector.detectMultiScale(sample_image)
# print bounding box for each detected face
for box in bboxes:
    print(box)
```

```python
# extract
for box in bboxes:
    x, y, width, height = box
    x2, y2 = x + width, y + height
# draw a rectangle over the pixels
    pixel = cv2.rectangle(sample_image, (x, y), (x2, y2), (0,0,255), 2)
```

```python
# show the image
cv2.imshow('face detection', pixel)
# keep the window open until we press a key
cv2.waitKey(0)
# close the window
cv2.destroyAllWindows()
```

```python
#visualize image with matplotlib
plt.imshow(cv2.cvtColor(pixel, cv2.COLOR_RGB2BGR))  
```
