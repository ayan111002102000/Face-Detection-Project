## Building model & Result Visualisation

```python
face_detector = MTCNN()
```

```python
face1 = face_detector.detect_faces(image1)
face2 = face_detector.detect_faces(image2)
face3 = face_detector.detect_faces(image3)
```

```python
def face_detector(image, face):
    plt.imshow(image)
    ax = plt.gca()
    for i in range(0, len(face)):
        x, y, width, height =  face[i]['box']
        rectangle = Rectangle((x, y), width, height, color = 'red', fill = False)
        ax.add_patch(rectangle)
    plt.show()
```

```python
face_detector(image1, face1)
```

```python
face_detector(image2, face2)
```

```python
face_detector(image3, face3)
```
